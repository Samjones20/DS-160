```python
car_sales = pd.DataFrame(data)
car_sales['Price'] = car_sales['Price'].str.replace('$', '').str.replace(',', '').astype('int64')
car_sales.head()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.head()
```

```python
car_sales["Sale Date"] = pd.date_range("12/2/2025", periods=len(car_sales))
car_sales.head()
```

```python
plt.plot(car_sales['Price'], car_sales['Total Sales'])
plt.title('Price vs Total Sales')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.title('Odometer (KM) vs Price')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

```python
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.title('Make vs Odometer (KM)')
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.show()
```

```python
plt.hist(car_sales['Make'])
plt.title('Histogram of Make')
plt.xlabel('Make')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(car_sales['Price'], car_sales['Total Sales'])
axs[0, 0].set_title('Price vs Total Sales')
axs[0, 0].set_xlabel('Price')
axs[0, 0].set_ylabel('Total Sales')

axs[0, 1].scatter(car_sales['Odometer (KM)'], car_sales['Price'])
axs[0, 1].set_title('Odometer (KM) vs Price')
axs[0, 1].set_xlabel('Odometer (KM)')
axs[0, 1].set_ylabel('Price')

axs[1, 0].bar(car_sales['Make'], car_sales['Odometer (KM)'])
axs[1, 0].set_title('Make vs Odometer (KM)')
axs[1, 0].set_xlabel('Make')
axs[1, 0].set_ylabel('Odometer (KM)')

axs[1, 1].hist(car_sales['Make'])
axs[1, 1].set_title('Histogram of Make')
axs[1, 1].set_xlabel('Make')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(x='Colour', y='Total Sales', data=car_sales)
plt.title('Box Plot of Colour and Total Sales')
plt.xlabel('Colour')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(100, 100))
sns.lmplot(x='Odometer (KM)', y='Total Sales', hue='Make', data=car_sales, ci=None)
plt.title('Relationship between Total Sales and Odometer (KM)')
plt.xlabel('Odometer (KM)')
plt.ylabel('Total Sales')
plt.show()
```

```python
sns.pairplot(car_sales)
plt.show()
```

```python
correlation = car_sales[['Total Sales','Price','Odometer (KM)']].corr()
correlation
```

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, linewidths=0.5, linecolor='black')
plt.title('Pearson Correlation Heatmap')
plt.show()
```
