```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("housing.csv")

print(data.head())
```

```python
categorical_cols = ['stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning']
for col in categorical_cols:
    data[col] = data[col].astype('category')
```

```python
# 1. Distribution of house prices
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Boxplots for categorical features vs. price
for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=data[col], y=data['price'])
    plt.title(f'House Price vs {col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 3. Pairplot for numerical features
numerical_cols = ['price', 'area', 'bedrooms', 'bathrooms']
sns.pairplot(data[numerical_cols])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
# 4. Correlation heatmap
plt.figure(figsize=(8, 6))
corr = data[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
